```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
team_data_file = "Golden State.csv"
salary_data_file = "Players_Salaries.xlsx"
```

```python
df_stats = pd.read_csv('Golden State.csv')
df_salary_2015 = pd.read_excel('Players_Salaries.xlsx', sheet_name='2015')
df_salary_2016 = pd.read_excel('Players_Salaries.xlsx', sheet_name='2016')
df_salary_2017 = pd.read_excel('Players_Salaries.xlsx', sheet_name='2017')
df_salary_2018 = pd.read_excel('Players_Salaries.xlsx', sheet_name='2018')
df_salary_2019 = pd.read_excel('Players_Salaries.xlsx', sheet_name='2019')
```

```python
df_stats.head()
```

```python
df_salary_2015.head()
```

```python
salary_df = pd.concat([
    df_salary_2015,df_salary_2016,df_salary_2017,df_salary_2018,df_salary_2019
],ignore_index=True,sort=False).reset_index(drop=True)
```

```python
salary_df
```

```python
name = salary_df['NAME'].str.split(pat=',',n=1,expand=True)
salary_df['NAME'] = name[0]
salary_df
```

```python
gs_stats_pd = pd.read_csv('Golden State.csv')
lakers_stats_pd = pd.read_csv('Lakers.csv')
laclippers_stats_pd = pd.read_csv('LA Clippers.csv')
sackings_stats_pd = pd.read_csv('Sacramento Kings.csv')
stats_df = pd.concat([
    gs_stats_pd,lakers_stats_pd,laclippers_stats_pd,sackings_stats_pd
],ignore_index=True,sort=False).reset_index(drop=True)
```

```python
stats_df
```

```python
merged_df = pd.merge(stats_df,salary_df, how='left', left_on=['PLAYER','YEAR'], right_on=['NAME','YEAR'])
merged_df
```

```python
merged_df.columns
```

```python
reduced_df = merged_df[['YEAR', 'PLAYER', 'TEAM_y', 'SALARY', 'AGE', 'GP', 'W', 'L', 'MIN', 'PTS', 'FGM',
       'FGA', 'FG%', '3PM', '3PA', '3P%', 'FTM', 'FTA', 'FT%', 'OREB', 'DREB',
       'REB', 'AST', 'TOV', 'STL', 'BLK', 'PF', 'FP', 'DD2', 'TD3', '+/-',
       'RK']]
```

```python
reduced_df
```

```python
final_df = reduced_df.rename(columns={
    "TEAM_y": "TEAM"
})
```

```python
final_df
```

```python
final_df = final_df.dropna(how="any")
```

```python
final_df
```

```python
final_df[["PLAYER", "3P%"]].head(30)
```

```python
#final_df.to_csv("final_df.csv", index=False, header=True)
```

```python
x_pts = final_df['PTS']
y_sal = final_df['SALARY']
plt.scatter(x_pts,y_sal,marker = 'o',edgecolors = 'black')
plt.title("NBA Points and Salaries")
plt.xlabel("Points")
plt.ylabel("Salary (10 Million)")
plt.grid()
plt.yticks(np.arange(0,50000000,10000000))
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_pts, y_sal)
regress_values = x_pts * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_pts,regress_values,"r-")
plt.annotate(line_eq,(-50,10),fontsize=15,color="red")
print(f"The r-squared is: {rvalue}")
```

```python
x_3p = final_df['3P%']
y_sal = final_df['SALARY']
plt.scatter(x_3p,y_sal,marker = 'o',edgecolors = 'black')
plt.title("NBA 3 Points Percentage and Salaries")
plt.xlabel("3P%")
plt.ylabel("Salary (10 Million)")
plt.grid()
plt.yticks(np.arange(0,50000000,10000000))
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_3p, y_sal)
regress_values = x_3p * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_3p,regress_values,"r-")
plt.annotate(line_eq,(-50,10),fontsize=15,color="red")
print(f"The r-squared is: {rvalue}")
```

```python
x_gp = final_df['GP']
y_sal = final_df['SALARY']
plt.scatter(x_gp,y_sal,marker = 'o',edgecolors = 'black')
plt.title("NBA Games Plaied and Salaries")
plt.xlabel("GP")
plt.ylabel("Salary (10 Million)")
plt.grid()
plt.yticks(np.arange(0,50000000,10000000))
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_gp, y_sal)
regress_values = x_gp * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_gp,regress_values,"r-")
plt.annotate(line_eq,(-50,10),fontsize=15,color="red")
print(f"The r-squared is: {rvalue}")
```
