```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
#load csv
final = ("../ClippersScripts_ha/highest_pd.csv")

highest_pd = pd.read_csv(final)
highest_pd
top_four = highest_pd.sort_values(by=['SALARY'], ascending=False)
top_four.head(20)
```

```python

x_pts = highest_pd['PTS']
y_sal = highest_pd['SALARY']
plt.scatter(x_pts,y_sal,marker = 'o',edgecolors = 'black')
plt.title("Number of Points and Highest Salaries in CA Teams")
plt.xlabel("Points")
plt.ylabel("Salary (10 Million)")
plt.grid()
plt.yticks(np.arange(0,50000000,10000000))
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_pts, y_sal)
regress_values = x_pts * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_pts,regress_values,"r-")
plt.show()
plt.annotate(line_eq,(-50,10),fontsize=15,color="red")
print(f"The r-squared is: {rvalue}")
```
