```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
#read correlation for Clippers file
clippers = ("../Project_1_Team_7/laclippers_corr.csv")

corr_clip_df = pd.read_csv(clippers,encoding='utf-8')
corr_clip_df.columns
corr_clip_df
```

```python
#rename column 'Unnamed: 0' to Column 1
renamed_df = corr_clip_df.rename(columns={"Unnamed: 0": "Column 1",
                                         "Unnamed: 1": "Column 2",
                                         "0": "Correlation"})
renamed_df
```

```python
#create dataframe of the correlation values of "SALARY" in Column 1
col_1_salary_df = renamed_df.loc[renamed_df["Column 1"]== "SALARY"]
col_1_salary_df
```

```python
#show only correlation over 0.5 in 'Column 1'
corr_1_strong = col_1_salary_df.loc[col_1_salary_df["Correlation"]>0.5,
                                   ["Column 1", "Column 2", "Correlation"]]
corr_1_strong
```

```python
#set labels, titles, and plot scatter
x_values = corr_1_strong['Column 2']
y_values = corr_1_strong['Column 1']
plt.scatter(x_values, y_values)
plt.xlabel('Stats')
plt.ylabel('Salary')
plt.show()
```
